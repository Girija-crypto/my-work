# filling null values

```python
df["Age"] = df.Age.fillna(df.Age.mean())      #filling null values using fillna method (filling mean of the row)
```

```python
df["Fare"] = df.Age.fillna(df.Age.median())     #filling median because it has outlayers
```

```python
df.Cabin.mode()     
```

```python
df["Cabin"] = df.Cabin.fillna(df.Cabin.mode()[0])      #filling mode because it is a categorical col  
```
